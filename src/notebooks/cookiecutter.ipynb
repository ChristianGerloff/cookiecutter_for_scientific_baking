### Versioning Control

At beginning of research projects, the organization of files, name conventions and also the analysis to be performed is often not formulated. In contrast, often the analysis is a process in which the analysis details and the organizational structure change constantly. This can make it very difficult to reproduce the analysis. Often this is due to a lack of version control.

#### Example

Do you have scripts on you computer looking like that :smiley:

     code/
        preprocessing_raw_20200321.mat
        badchannels_v1_20200321.mat
        test/
           final_2/
              preprocessing_raw_fixed_v3.mat
              preprocessing_raw_v4.mat
        final/
           archive/
              preprocessing_raw_fixed.mat    
           preprocessing_raw.mat
             badchannels_v1_20200321.mat


Do you know what is the current version, what was the last version which you used to provide data to your team mate? 

And do you know what changed during the versions?

***Summarized:***
- Your chance to ensure the quality of your code and consistency of your results without versioning control is probably very low

## Example reproducable code

```python
import sys
import numpy as np
import pandas as pd
import seaborn as sns
sys.path.append('../model')
import splithalf as sp
```

```python
observations = pd.DataFrame(np.random.randint(low=6,high=50, size=(100, 2)), columns=['age','hopefully_not_iq'])
ax = sns.violinplot(x=observations['age'])
```

```python
s1_subsample, s2_subsample = sp.simple_split_half(observations, split=0.5)
display(sp.pearson(s1_subsample,s2_subsample, cols=['age','hopefully_not_iq']))
ax = sns.violinplot(data=pd.concat([s1_subsample,s2_subsample]), x='subsample', y='age')
```

```python
s1_subsample, s2_subsample = sp.simple_split_half(observations, split=0.5)
display(sp.pearson(s1_subsample,s2_subsample, cols=['age','hopefully_not_iq']))
ax = sns.violinplot(data=pd.concat([s1_subsample,s2_subsample]), x='subsample', y='age')
```

# Reproducable environment

Every package / toolbox / framework provides you with some functions / processes without you having to deal with the details under the hood. Consequently, each toolbox is based on some assumptions made in the toolbox, such as the way it is implemented.

This becomes clear when a new version of a toolbox or, for example, a new version of Matlab is released and your own code stops working, or worse, when the results change unnoticed.
